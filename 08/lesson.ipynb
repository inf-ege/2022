###### Исполнитель Редактор

Исполнитель Редактор получает на&nbsp;вход строку цифр и&nbsp;преобразовывает&nbsp;её. Редактор может выполнять две команды, в&nbsp;обеих командах v&nbsp;и&nbsp;w&nbsp;обозначают цепочки цифр.

А) **заменить (v, w)**. Эта команда заменяет в&nbsp;строке первое слева вхождение цепочки v&nbsp;на&nbsp;цепочку&nbsp;w. Например, выполнение команды **заменить (111, 27)** преобразует строку **05111150** в&nbsp;строку **0527150**.

Если в&nbsp;строке нет вхождений цепочки&nbsp;v, то&nbsp;выполнение команды **заменить (v, w)** не&nbsp;меняет эту строку.

Б) **нашлось (v)**. Эта команда проверяет, встречается&nbsp;ли цепочка v&nbsp;в&nbsp;строке исполнителя Редактор. Если она встречается, то&nbsp;команда возвращает логическое значение &laquo;истина&raquo;, в&nbsp;противном случае возвращает значение &laquo;ложь&raquo;. Строка исполнителя при этом не&nbsp;изменяется.

Цикл
```
ПОКА условие
  последовательность команд
КОНЕЦ ПОКА
```
выполняется, пока условие истинно.

В&nbsp;конструкции
```
ЕСЛИ условие
  ТО команда1
  ИНАЧЕ команда2
КОНЕЦ ЕСЛИ
```
выполняется команда1 (если условие истинно) или команда2 (если условие ложно).

Ниже приведена программа для исполнителя Редактор.

**Задание 12.5** 
```
НАЧАЛО
    ПОКА нашлось (19) ИЛИ нашлось (299) ИЛИ нашлось (3999)
      заменить (19, 2)
      заменить (299, 3)
      заменить (3999, 1)
    КОНЕЦ ПОКА
КОНЕЦ
``` 
На&nbsp;вход этой программе подаётся строка длины 101, состоящая из&nbsp;цифры&nbsp;2, за&nbsp;которой следуют 100 идущих подряд цифр 9. Какая строка получится в&nbsp;результате применения программы к&nbsp;этой строке?

````python
s = '2' + '9' * 100

while '19' in s or '299' in s or '3999' in s:
    s = s.replace('19', '2', 1)
    s = s.replace('299', '3', 1)
    s = s.replace('3999', '1', 1)

print(s)
````

**Задание 12.6**
```
НАЧАЛО
ПОКА нашлось(111) или нашлось(333)
  ЕСЛИ нашлось(111) 
  ТО заменить(111, 3)
  ИНАЧЕ заменить(333, 1)
КОНЕЦ ПОКА
КОНЕЦ
```
На&nbsp;вход программе подана строка из&nbsp;более чем 100 подряд идущих символов &laquo;3&raquo;. Найдите минимальную длину входной строки, в&nbsp;результате обработки которой получится минимальное возможное число.

````python
min_len = 500
min_num = int('3' * min_len)
for x in range(101, 300):
    s = '3' * x
    while '111' in s or '333' in s:
        if '111' in s:
            s = s.replace('111', '3', 1)
        else:
            s = s.replace('333', '1', 1)
    if int(s) < min_num:
        min_num = int(s)
        min_len = x
print(min_len)
````

**Задание 24.3** Текстовый файл состоит не&nbsp;более чем из&nbsp;$10^6$ символов X, Y&nbsp;и&nbsp;Z.&nbsp;Определите максимальное количество идущих подряд символов, среди которых каждые два соседних различны.

````python
file = open('24.txt').read()

for x in 'QWERTYUIOPASDFGHJKLZXCVBNM':
    while x+x in file:
        file = file.replace(x+x, x+' '+x)

file = file.split()
maxL = max(file, key=len)
print(len(maxL))
````

**Задание 24.4** Текстовый файл 24.4.txt содержит строку из&nbsp;набора A, B, C, D, E, F, всего не&nbsp;более чем из&nbsp;$10^6$&nbsp;символов. Найдите максимальное количество подряд идущих пар символов АB, CB, BС&nbsp;и&nbsp;BA, стоящих одна за&nbsp;другой и&nbsp;пересекающихся с&nbsp;соседними парами одной буквой. Например, в&nbsp;строке BDE**ABCBAB**BD такие пары составляют подстроку ABCBAB = AB + BC + CB + BA + AB, итого 5&nbsp;пар.

<!-- ![](img/24/24.4.png) -->

````python
file = open('24.4.txt').read()

for d in 'DEF':
    file = file.replace(d, ' ')

for a, c in 'AC', 'CA', 'AA', 'BB', 'CC':
    while a+c in file:
        file = file.replace(a+c, a+' '+c)

file = file.split()
maxL = max(file, key=len)
print(len(maxL)-1)
````

**Задание 24.7** Текстовый файл 24.7.txt состоит не&nbsp;более чем из&nbsp;$10^6$ символов и&nbsp;содержит только заглавные буквы латинского алфавита (ABC...Z). Текст разбит на&nbsp;строки различной длины. 

Необходимо найти строку, содержащую наибольшее количество букв&nbsp;Q (если таких строк несколько, надо взять&nbsp;ту, которая в&nbsp;файле встретилась позже). Определите, какая буква встречается в&nbsp;этой строке реже всего (но&nbsp;присутствует!). Если таких букв несколько, надо взять&nbsp;ту, которая стоит раньше в&nbsp;алфавите. Запишите в&nbsp;ответе эту букву, а&nbsp;затем&nbsp;&mdash; сколько раз она встречается во&nbsp;всем файле.

Пример. Исходный файл:
```
ZZQAQB
QAVQAB
BAQTUB
```
В&nbsp;этом примере в&nbsp;первой и&nbsp;второй строках по&nbsp;две букву&nbsp;Q, в&nbsp;третьей&nbsp;&mdash; одна. Берём вторую строку, т.к. она стоит в&nbsp;файле позже. В&nbsp;этой строке реже других встречаются буквы&nbsp;V и&nbsp;B (по&nbsp;одному разу), выбираем букву&nbsp;B, т.&nbsp;к. она раньше стоит в&nbsp;алфавите. В&nbsp;ответе для этого примера надо записать&nbsp;B4, так как во&nbsp;всех строках файла буква&nbsp;B встречается 4&nbsp;раза.

````python
file = open('24.7.txt').readlines()
lineQ = ''
countQ = float('-inf')

for line in file:
    Q = line.count('Q')
    if Q >= countQ:
        lineQ = line
        countQ = Q

minChar = float('inf')
saveChar = ''

for char in sorted('QWERTYUIOPASDFGHJKLZXCVBNM'):
    count = lineQ.count(char)
    if count > 0 and count < minChar:
        minChar = count
        saveChar = char

print(saveChar, minChar, sep='')
````
