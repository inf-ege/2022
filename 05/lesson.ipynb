## Задания 19–21. Одна куча

Два игрока, Петя и&nbsp;Ваня, играют в&nbsp;следующую игру. Перед игроками лежит куча камней. Игроки ходят по&nbsp;очереди, первый ход делает Петя. За&nbsp;один ход игрок может добавить в&nbsp;кучу один или четыре камня либо увеличить количество камней в&nbsp;куче в&nbsp;пять раз. Например, имея кучу из&nbsp;15&nbsp;камней, за&nbsp;один ход можно получить кучу из&nbsp;16, 19&nbsp;или 75&nbsp;камней. У&nbsp;каждого игрока, чтобы делать ходы, есть неограниченное количество камней. Игра завершается в&nbsp;тот момент, когда количество камней в&nbsp;куче становится не&nbsp;менее 68.

Победителем считается игрок, сделавший последний ход, т. е. первым получивший кучу, в&nbsp;которой будет 68&nbsp;или больше камней. В&nbsp;начальный момент в&nbsp;куче было&nbsp;S камней; 1 &le; S &le; 67.

Будем говорить, что игрок имеет выигрышную стратегию, если он&nbsp;может выиграть при любых ходах противника. Описать стратегию игрока &mdash; значит, описать, какой ход он&nbsp;должен сделать в&nbsp;любой ситуации, которая ему может встретиться при различной игре противника. В&nbsp;описание выигрышной стратегии не&nbsp;следует включать ходы играющего по&nbsp;этой стратегии игрока, не&nbsp;являющиеся для него безусловно выигрышными, т. е.&nbsp;не&nbsp;являющиеся выигрышными независимо от&nbsp;игры противника.

**Задание 19** Известно, что Ваня выиграл своим первым ходом после неудачного первого хода Пети. Укажите минимальное значение&nbsp;S, когда такая ситуация возможна.

```python
def g(a, p):
    if a >= 68 and p == 2:
        return True
    if a >= 68 or p == 2:
        return False
    return g(a+1, p+1) or g(a+4, p+1) or g(a*5, p+1)

for S in range(1, 67+1):
    if g(S, 0):
        print(S)
        break
```

## Задания 19–21. Без повтора

За&nbsp;один ход игрок может:

– добавить в&nbsp;кучу три камня (действие&nbsp;Б)  
– или утроить количество камней в&nbsp;куче, а&nbsp;затем убрать из&nbsp;кучи один камень (действие В). 

Например, имея кучу из&nbsp;10&nbsp;камней, за&nbsp;один ход можно получить кучу из&nbsp;11, 13&nbsp;или 29&nbsp;камней. У&nbsp;каждого игрока, чтобы делать ходы, есть неограниченное количество камней. 

**Игрокам запрещается повторять предыдущий ход соперника.**

Игра завершается в&nbsp;тот момент, когда количество камней в&nbsp;куче становится более 32. Победителем считается игрок, сделавший последний ход, то&nbsp;есть первым получивший кучу, в&nbsp;которой будет 33&nbsp;или больше камней.
```python
def g(a, p, h):
    if a > 32 and p == 2: return True
    if a > 32 or p == 2: return False
    if h == 1:
        return g(a+3, p+1, 2) or g(a*3-1, p+1, 3)
    if h == 2:
        return g(a+1, p+1, 1) or g(a*3-1, p+1, 3)
    if h == 3:
        return g(a+1, p+1, 1) or g(a+3, p+1, 2)
    return g(a+1, p+1, 1) or g(a+3, p+1, 2) or g(a*3-1, p+1, 3)
    
for S in range(1, 32+1):
    if g(S, 0, 0):
        print(S)
        break
```
